```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, desc, sum, avg
# Initialize
spark = SparkSession.builder.appName("SparkCompleteNotes").getOrCreate()
# Create Base DataFrame
data = [
    (1, "Alice", "Engineering", 75000, 25),
    (2, "Bob", "Marketing", 60000, 30),
    (3, "Charlie", "Engineering", 80000, 35),
    (4, "David", "Sales", 65000, 28),
    (5, "Eve", "Marketing", 70000, 32)
]
columns = ["Id", "Name", "Department", "Salary", "Age"]
df = spark.createDataFrame(data, columns);
```

```python
df_select = df.select("Name", "Department", "Salary")

df_select.show()
```

```python
df_filter=df.filter(col("salary")>75000)
df_filter.show()
```

```python
df_filter.show()
```

```python
df.write.mode("overwrite").csv("output/employeesTable",header=True) 
```

```python
df.toPandas().to_csv(
"output/employees123.csv",
    index=False)
```

```python
df.createOrReplaceTempView("employees")
```

```python
optimized_df = df.filter(col("Salary") > 500).filter(col("Id") < 500000).select("Id", "Salary")
```

```python
optimized_df.explain()
```

```python
import time

start_time = time.time()
count_uncached = optimized_df.count()  # action triggers dag
end_time = time.time()
print(f"1.optimized execution | count:{count_uncached} | Time:{round(end_time-start_time,4)} seconds")
```

```python
optimized_df.cache()
```

```python
start_time = time.time()
count_second_cache = optimized_df.count()
end_time = time.time()
print(f"3. SECOND CACHED Action | Count: {count_second_cache} | Time: {round(end_time - start_time, 4)} seconds")
```

```python
data = [("Alice", 25), ("Bob",17),("Charlie", 35),("David", 15)]
df1 = spark.createDataFrame(data, ["Name", "Age"])
                                                      #udf
df1.show()
```

```python
def categorize_age(age):
    if age>=18:
        return "Adult"
    return "Minor"
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

age_category_udf = udf(categorize_age, StringType())

df1_with_category = df1.withColumn("age_category", age_category_udf(col("Age")))
df1_with_category.show() 
```

```python
spark.udf.register("sql_categorize_age", categorize_age, StringType())
df1.createOrReplaceTempView("people")
```

```python
sql_df = spark.sql("""
SELECT
    Name,
    Age,
    sql_categorize_age(Age) AS Category
FROM people
""")
sql_df.show()     
```
